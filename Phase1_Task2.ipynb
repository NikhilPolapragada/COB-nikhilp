```python
# HANDLING MISSING VALUES
# Since the feature data['director'] has a lot of missing values( around 30%), it is best to remove it from the dataset
```

```python
data=data.drop(columns='director')
data
```

```python
#Since,the 200 data points of df['country'] is missing, those missing data points will be removed
```

```python
data=data.drop(data[data['country']=='Not Given'].index)
data
```

```python
# REMOVING OUTLIERS
```

```python
data['country'].value_counts()
countries=dict()
s=0
for i in data.country:
    if i not in countries:
        countries[i]=1
    else:
        countries[i]+=1
countries
```

```python
#finding 10% outlier data to be removed
minimum=[]
Sum=0
for i in countries:
    if countries[i]>=1 and countries[i]<=80:
     minimum.append(i)
     Sum+=countries[i]
print(Sum,len(minimum))  
```

```python
for i in minimum:
    data=data.drop(data[data['country']==i].index)
data
```

```python
# CHECKING FOR CORELATION BETWEEN FEATURES
# the features which are possible to be labelled will be labelled using label encoder
```

```python
from sklearn.preprocessing import LabelEncoder
encode=LabelEncoder()

#labelling type feature
encode.fit(data.type)
new_type=encode.transform(data.type)
data.type= new_type

#labelling country feature
encode.fit(data.country)
new_country=encode.transform(data.country)
data.country= new_country

#labelling date feature
encode.fit(data.date_added)
new_date=encode.transform(data.date_added)
data.date_added= new_date

#labelling rate feature
encode.fit(data.rating)
new_rate=encode.transform(data.rating)
data.rating= new_rate

#labelling duration feature
encode.fit(data.duration)
new_dur=encode.transform(data.duration)
data.duration= new_dur

#labelling show_id feature
encode.fit(data.show_id)
new_id=encode.transform(data.show_id)
data.show_id= new_id
```
